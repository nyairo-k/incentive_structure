## Question 3: We would like to design a compensation structure for the agents that specifically targets agents who are close to the best agents, but are not quite the best yet. Really, we’re trying to motivate them to step up at the edge of where we see big performance transitions.  We’d like the total average compensation for a task to be at most 300KSH across all agents. We’d like every task to have a minimum payment of at least 175KSH so agents always feel like they can cover transport costs. Please design an incentive structure for agents in the form of a per task rate and a weekly performance bonus based on the tasks completed in that week.
Average compensation per task 300 at most, 175 at least for **each** agent

We need a summary of all tasks completed by each agent in week

Bonus will be given to agents who hit a particular number of tasks done in a week.

Code below gives us a dataframe containing the total task count for each agent in a week

```python
#group data into weekly segments(84/7)
weekly_tasks = data.groupby(data.index // 7).sum()
weekly_tasks.index = [f'Week_{i+1}' for i in range(weekly_tasks.shape[0])]
weekly_tasks
```

```python
# Overview stats
min_per_week = weekly_tasks.values.flatten().min()
max_per_week = weekly_tasks.values.flatten().max()
std_per_week = weekly_tasks.values.flatten().std()
mean_per_week = weekly_tasks.values.flatten().mean()
print("highest number of tasks completed per week:", max_per_week)
print("lowest number of tasks completed per week:", min_per_week)
print("standard deviation per week:", std_per_week)
print("Meanscore per week:", mean_per_week)
```

From the above stats we see that there are agents completing up to 110 tasks in a week and some 0 with a significant high variability.

How do we reward these best perfomers and motivate the rest to step up the edge? How do we compensate the agents according to the number of tasks each agent has completed.

Given the high variations of our dataset, we have multiple ranges within which a data point might fall. This ranges will help us define our incentive structure as follows:

1. Low Performers: 0-10 tasks/week

2. Moderate Performers: 11-30 tasks/week

3. Above Average Performers: 31-70 tasks/week

4. High Performers: 71-100 tasks/week

5. Top Performers: 101+ tasks/week

```python
#Now categorize task count into ranges per week
ranges = [(0, 10), (11, 30), (31, 70), (71, 100), (101, 200)]
def categorize_task_counts(row, ranges):
    categorized = {}
    for task_range in ranges:
        # Count how many agents fall into each range
        count = ((row >= task_range[0]) & (row <= task_range[1])).sum()
        categorized[f'{task_range[0]}-{task_range[1]}'] = count
    return categorized
```

```python
# Apply this function to each week (each row in the DataFrame)
weekly_range_counts = weekly_tasks.apply(lambda row: categorize_task_counts(row, ranges), axis=1)
```

```python
# Convert to DataFrame for easier interpretation
weekly_range_counts_df = pd.DataFrame(list(weekly_range_counts))
weekly_range_counts_df.index = weekly_tasks.index
print(weekly_range_counts_df)
```

```python
#Visualizing tasks completed by agents every week
custom_colors = ['orange', 'green', 'skyblue', 'purple', 'green']
weekly_range_counts_df.plot(kind='bar', stacked=True, figsize=(10, 6), color= custom_colors)
plt.title('Task Range Distribution Across Weeks')
plt.xlabel('Week')
plt.ylabel('Number of Agents')
plt.xticks(rotation=45)
plt.legend(title='Task Range', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.tight_layout()
plt.show()
```

## Findings
From the above analysis,  we see that the distribution of agents across the task ranges remains fairly consistent over the 12 weeks.
There's a high concentration of low performers with majority of agents falling into the 0-10 tasks range. For example, in Week 1, 666 agents completed only 0-10 tasks, and this trend continues throughout the weeks, with little variation.This indicates a large group of agents consistently performing at a very low level. 

We see consinstent middle range(31-70 range) performers, with between 471 and 537 agents completing this number of tasks each week and a smaller but consistent number of agents completing tasks in the 11-30 range each week.

We see a ceiling effect weeks 6, 8, 9, 10, 11, 12 with a few agents performing tasks ranging 71-200.

Given the findings above my base compensation structure will be as follows:

Tier 0 **(Low Performers (0-10 tasks))**: 175 KSH/task

Tier 1 **(Moderate Performers (11-30 tasks))**: 200 KSH/task

Tier 2 **(Above Average Performers (31-70 tasks))**: 225 KSH/task

Tier 3 **(High Performers (71-100 tasks))**: 275 KSH/task

Tier 4 **(Top Performers (101+ tasks))**: 300 KSH/task

Most agents fall within the low performers category and above average performers category. As an incentive, the base compensation increment rate for above average perfomers is much more to motivate more workers to move on to the high performers tier.
